```python
import pandas as pd
df=pd.read_csv(r'E:\datasets\dis-sym.csv')
df
```

```python
# forming a new col : have list of all diseases

df["list_of_Symps"] = 0
for i in range(df.shape[0]):
    values = df.iloc[i].values
    values = values.tolist()
    if 0 in values:
        df["list_of_Symps"][i] = values[1:values.index(0)]
    else:
        df["list_of_Symps"][i] = values[1:]

df
```

```python
# total no. of diseases

print(len(list(df.Disease.unique())))
df.Disease.unique()
```

```python
# total no. of symptoms

a=set()

for i in df['list_of_Symps']:
    for j in i:
        a.add(j)

a=list(a)
print(a)

len(a)
```

```python
# Function to convert list to set, ignoring non-string elements

def list_to_set(symptom_list):
    return list({symptom for symptom in symptom_list if isinstance(symptom, str)})

# Apply the function to the 'list_of_Symps' column
df['set_of_Symps'] = df['list_of_Symps'].apply(list_to_set)

df
```

```python
# Drop rows with duplicate sets in 'set_of_Symps' column

df = df.drop_duplicates(subset='set_of_Symps')
df
```

```python
# total no. of diseases

print(len(list(df.Disease.unique())))
df.Disease.unique()
```

```python
# drop the "list_of_Symps" col

df = df.drop(columns=['list_of_Symps'])
df = df.reset_index(drop=True)
df
```
