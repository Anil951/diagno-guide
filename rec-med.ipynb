```python
# Rx_OTC

# Rx-to-OTC switch is the transfer of proven prescription drugs to nonprescription, where
# OTC (Over-the-counter) = Medication that can be purchased without a medical prescription
# Rx = Prescription Needed
# Rx/OTC = Prescription or Over-the-counter.

# pregnancy:

# A- no risk at first trimester of pregnancy by that drug
# B- by Animal studies:no risk on fetus (pregnant women)--no sufficient studies conducted in humans. 
# C- by Animal studies:risk on fetus but can use for potential benefits --no sufficient studies conducted in humans. 
# D- by Human studies: risk on fetus but can use for potential benefits
# X- Should not use by pregnant women
# N- not been classified by the FDA in terms of its safety during pregnancy

# PRIORITY(from safe to harmful):
#     A-B-C/D-N-X 
    
# alchohol:
    
# X- interacts with alcohol
# N -no alcohol interaction
```

```python
from fuzzywuzzy import fuzz

temp=[]
def print_similar_elements(list1, list2, similarity_threshold,c=0):

    for elmt1 in list1:
        for elmt2 in list2:
            token_set_ratio = fuzz.token_set_ratio(elmt1, elmt2)
            token_sort_ratio = fuzz.token_sort_ratio(elmt1, elmt2)
            wratio = fuzz.WRatio(elmt1, elmt2)

            # Return the maximum similarity score
            similarity_score=max(token_set_ratio, token_sort_ratio, wratio)
            if similarity_score > similarity_threshold:
                c+=1
                temp.append(elmt1)
                print(elmt1, "--", elmt2)
                break  # Break out of the inner loop after finding a match
    print("count:",c)
# Print similar elements with similarity above 70
similarity_threshold = 75
print_similar_elements(list2, list1, similarity_threshold,c=0)

print(temp)

```
